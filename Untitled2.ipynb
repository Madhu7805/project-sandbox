```python
import cv2
import matplotlib.pyplot as plt

image_path = "/content/itzz mee.jpg"

# Check if the image file exists
import os
if not os.path.exists(image_path):
    raise FileNotFoundError(f"Image file not found at: {image_path}")

# Load the image
image = cv2.imread(image_path)

# Check if image loading was successful
if image is None:
    raise ValueError(f"Failed to load image from: {image_path}")

# Corrected function name from 'cvtcolor' to 'cvtColor'
rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

scale_factor = 0.8
for i in range(5):
    position = [1, 2, 3, 4, 5]
    new_width = int(rgb_img.shape[1] * (scale_factor ** i))
    new_width = int(rgb_img.shape[0] * (scale_factor ** i))

    # Corrected function name from 'cv2resize' to 'cv2.resize'
    resized_image = cv2.resize(rgb_img, (new_width, new_width), interpolation=cv2.INTER_AREA)

    plt.subplot(3, 3, position[i])  # Use index to access position values
    plt.imshow(resized_image)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
