```python
import cv2
import os
from google.colab import files
from google.colab.patches import cv2_imshow # Import cv2_imshow


def generate_gaussian_pyramid(image, levels):
    pyramid = [image]
    for i in range(levels - 1):
        downsampled_image = cv2.pyrDown(pyramid[-1])
        pyramid.append(downsampled_image)
    return pyramid

# Upload the image to Colab
# uploaded = files.upload()  # If you need to upload, uncomment this line
# image_path = list(uploaded.keys())[0]  # Use this if you upload with files.upload()
image_path ="/content/itzz mee.jpg" # uploaded is already a string with the path

# Check if the image was loaded successfully
image = cv2.imread(image_path)
if image is None:
    print(f"Error: Could not load image from {image_path}")
else:
    pyramid = generate_gaussian_pyramid(image, 5)

    for i, level in enumerate(pyramid):
        cv2_imshow(level) # Use cv2_imshow instead of cv2.imshow
        print(f"Level {i}") # Add a print statement to indicate the level
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
